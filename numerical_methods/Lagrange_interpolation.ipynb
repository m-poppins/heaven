Интерполировать таблично заданную функцию, используя полином Лагранжа 4 порядка

```python
import matplotlib.pyplot as plt


def inpol(x, y, n):
    def polynomial(x0):
        begin = 0
        for f in x:
            if f > x0:
                break
            elif f == x0:
                return y[begin]
            begin += 1
        begin = min(max(0, begin - int(n/2 + 1)), len(x) - 1 - n)
        result = 0
        for k in range(n+1):
            r = y[begin + k]
            for m in range(n+1):
                if m != k:
                    r *= (x0 - x[begin+m]) / (x[begin+k] - x[begin+m])
            result += r
        return result

    res_x = [x[0]]
    for i in range(len(x)-1):
        d = (x[i+1] - x[i]) / 4
        res_x += [x[i] + d, x[i] + 2*d, x[i] + 3*d, x[i+1]]
    return res_x, [polynomial(dot) for dot in res_x]


_x = [i+1 for i in range(20)]
_y = [5, 6, 8, 10, 12, 13, 12, 10, 8, 10, 8, 11, 7, 9, 11, 10, 9, 12, 11, 6]
_res_x, _res_y = inpol(_x, _y, 4)
print(_res_x ,"\n")
print(_res_y)


grid_size = (2, 1)
fig = plt.figure(figsize=(16, 8))
#ax1 = plt.subplot2grid(grid_size, (0, 0), colspan=2)
ax2 = plt.subplot2grid(grid_size, (1, 0), colspan=2)
#ax1.plot(_x, _y, 'mD--')
ax2.plot(_res_x, _res_y, 'bo-',color="#900000",markersize=4)
plt.show()

```
